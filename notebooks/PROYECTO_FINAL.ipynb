4.2. Conversión de variables

a)Convertir las columnas str a fecha

```python
columnas_fecha = [
    "last_scraped",
    "host_since",
    "calendar_last_scraped",
    "first_review",
    "last_review"
]

for col in columnas_fecha:
    listado[col] = pd.to_datetime(listado[col])
```

```python
listado[columnas_fecha].dtypes
```

b) Convertir la columna PRICE a formato numérico

```python
listado["price"] = (
    listado["price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
listado["price"].describe()
```

c)Cambio de str a porcentajes

```python
listado["host_response_rate"] = (
    listado["host_response_rate"]
    .str.replace("%", "", regex=False)
    .astype("float")
)

listado["host_acceptance_rate"] = (
    listado["host_acceptance_rate"]
    .str.replace("%", "", regex=False)
    .astype("float")
)
```

```python
listado[
    ["host_response_rate", "host_acceptance_rate"]
].describe()
```

d) Cambio de str a booleanos

```python
columnas_bool = [
    "host_is_superhost",
    "host_has_profile_pic",
    "host_identity_verified",
    "instant_bookable"
]

for col in columnas_bool:
    listado[col] = listado[col].map({
        "t": True,
        "f": False
    })
```

```python
listado[columnas_bool].dtypes
```

Al no haberse convertido todo correctamente a bool es porque esas columnas tienen más valores a parte de t y f. Comprobamos:

```python
for col in [
    "host_is_superhost",
    "host_has_profile_pic",
    "host_identity_verified"
]:
    print(f"\n{col}")
    print(listado[col].unique())
```

Salen las opciones True, False y nan. Cambiarlas al booleano que acepte nan

```python
columnas_bool = [
    "host_is_superhost",
    "host_has_profile_pic",
    "host_identity_verified"
]

for col in columnas_bool:
    listado[col] = listado[col].astype("boolean")
```

```python
listado[columnas_bool].dtypes
```

4.2. Barrios más caros de Madrid:

```python
alojamientos = dataset_eda.drop_duplicates(subset="id")

alojamientos.groupby(
    "neighbourhood_cleansed"
)["price"].agg(["count","mean"]).sort_values(
    "mean",
    ascending=False
)
```

Barrios con minimo 10 alojamientos:

```python
barrios_precio = (
    alojamientos.groupby("neighbourhood_cleansed")["price"]
    .agg(["count","mean"])
)

barrios_precio = barrios_precio[barrios_precio["count"] >= 10]

barrios_precio.sort_values("mean", ascending=False)
```

4.3. Los Superhosts cobran más que los anfitriones normales?

```python
alojamientos.groupby(
    "host_is_superhost"
)["price"].agg(["count","mean"])
```

4.4. Habitaciones y precio. ¿Más habitaciones = más precio?

```python
alojamientos.groupby(
    "bedrooms"
)["price"].agg(["count","mean"]).sort_values("mean")
```

4.5. Tipo de alojamiento vs precio

```python
alojamientos.groupby(
    "property_type"
)["price"].agg(["count","mean"]).sort_values(
    "mean",
    ascending=False
).head(15)
```

4.6. Valoraciones vs precio

```python
alojamientos["grupo_valoracion"] = pd.cut(
    alojamientos["review_scores_rating"],
    bins=[0, 4, 4.5, 4.8, 5],
    labels=["Baja", "Media", "Alta", "Excelente"]
)
```

```python
alojamientos.groupby("grupo_valoracion")["price"].agg(["count","mean"])
```
